### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep="\t")
chipo
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price'] = chipo['item_price'].replace('[$]',"",regex=True).astype(float)
product_over_10 = chipo['item_price'][chipo['item_price']>10.00]
print("Number of products costing more than $10.00:", len(product_over_10))
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
chipo[['item_name','item_price']].drop_duplicates()
```

### Step 6. Sort by the name of the item

```python
df_sort = chipo.sort_values(by="item_name")
df_sort.set_index('item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
loc = chipo['item_price'].idxmax()
most = chipo.iloc[loc]['quantity']
print(f"Quantity of the most expensive item ordered : {most}")
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie = chipo['item_name'][chipo['item_name']=='Veggie Salad Bowl']
print(f"Number of times Veggie Salad Bowl was ordered : {len(veggie)}")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
soda = chipo[chipo['item_name'] == 'Canned Soda']
total = soda[soda['quantity'] >= 2].count()[0]
print(f"Number of times more than one Canned Soda was ordered : {total}")
```
