### Step 3. Use the tsv file and assign it to a dataframe called food

```python
import pandas as pd
food = pd.read_csv("/kaggle/input/world-food-facts/en.openfoodfacts.org.products.tsv",sep="\t")
```

```python
food.head()
```

### Step 5. What is the number of observations in the dataset?

```python
num_observations = food.shape[0]
print(f"Number of observations is {num_observations}")
```

### Step 6. What is the number of columns in the dataset?

```python
num_columns = food.shape[1]
print(f"Number of columns is {num_columns}")
```

### Step 7. Print the name of all the columns.

```python
name_columns = food.columns.to_list()
print(f"Names of all the columns is \n{name_columns}")
```

### Step 8. What is the name of 105th column?

```python
name_of_105th = food.columns[104]
print(f"Name of of 105th column is {name_of_105th}")
```

### Step 9. What is the type of the observations of the 105th column?

```python
type_105th_column = food.iloc[:, 104].dtype
print(f"Data type of observations in the 105th column is {type_105th_column}")
```

### Step 10. How is the dataset indexed?

```python
indexed = food.index
print(f"Dataset indexed : {indexed}")
```
