### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.offline import iplot,plot
from plotly.subplots import make_subplots
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv")
titanic
```

```python
titanic.sample(5)
```

### Step 4. Set PassengerId as the index

```python
titanic.set_index('PassengerId',inplace=True)
titanic
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
gender = titanic['Sex'].value_counts()
iplot(px.pie(gender,
             values="Sex",
             title='Male/Female Proportion',
             names=['Male','Female'],
             labels={'Sex':'Number of Person'}
).update_traces(textinfo='value+percent'))
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
iplot(px.scatter(titanic,
                 x='Age',
                 y='Fare',
                 color='Sex',
                 title='Fare vs Age by Gender'
))
```

### Step 7. How many people survived?

```python
survived = titanic['Survived'].value_counts().reset_index().rename(columns={'index':'gender'}).replace({0:'Male',1:'Female'}).set_index('gender')

fig = px.bar(survived,
             x=survived.index,
             y='Survived',
             text_auto=True,
             labels={'Survived': 'Count'},
             color_discrete_sequence=[ '#7a0765'],
             title='Count of Survivors by Gender')

fig.show()
```

### Step 8. Create a histogram with the Fare payed

```python
fig = px.histogram(titanic, x='Fare', title='Distribution of Fare Paid')
fig.show()
```
