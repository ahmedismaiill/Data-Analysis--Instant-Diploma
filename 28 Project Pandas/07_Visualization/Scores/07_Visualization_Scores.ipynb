### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.offline import iplot , plot
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'female': [0, 1, 1, 0, 1],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
iplot(px.scatter(df,x='preTestScore',y='postTestScore',size='age',labels={'preTestScore': 'Pre-Test Score', 'postTestScore': 'Post-Test Score'},title='Scatterplot of Pre-Test Score and Post-Test Score with Size Based on Age'))
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
df['female'] = df['female'].replace({0: 'male', 1: 'female'})
iplot(px.scatter(df,
                 x='preTestScore',
                 y='postTestScore',
                 size='age',
                 color='female',
                  title='Scatterplot of Pre-Test Score and Post-Test Score with Size Based on Age',
                 color_discrete_map={0: 'blue', 1: 'pink'}

                 ))
```
