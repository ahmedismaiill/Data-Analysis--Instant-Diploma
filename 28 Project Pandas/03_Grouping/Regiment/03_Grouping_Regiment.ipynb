### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Create the DataFrame with the following values:

```python
raw_data = {'regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
        'company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd','1st', '1st', '2nd', '2nd'],
        'name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
        'preTestScore': [4, 24, 31, 2, 3, 4, 24, 31, 2, 3, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70, 25, 94, 57, 62, 70, 62, 70]}
```

### Step 3. Assign it to a variable called regiment.
#### Don't forget to name each column

```python
regiment = pd.DataFrame(raw_data)
regiment
```

### Step 4. What is the mean preTestScore from the regiment Nighthawks?  

```python
reg_mean = regiment.set_index('regiment')
reg_mean = reg_mean.loc['Nighthawks']['preTestScore'].mean()
print(f"Mean preTestScore for Nighthawks regiment : {reg_mean}")
```

### Step 5. Present general statistics by company

```python
general_statics = regiment.groupby('company').describe()
general_statics
```

### Step 6. What is the mean of each company's preTestScore?

```python
testscore_company = regiment.groupby('company')['preTestScore'].mean()
pd.DataFrame(testscore_company)
```

### Step 7. Present the mean preTestScores grouped by regiment and company

```python
pretest_regiment_company = regiment.groupby(['regiment', 'company'])['preTestScore'].mean()
pd.DataFrame(pretest_regiment_company)
```

### Step 8. Present the mean preTestScores grouped by regiment and company without heirarchical indexing

```python
pretest_regiment_company = regiment.groupby(['regiment', 'company'])['preTestScore'].mean().reset_index()
pretest_regiment_company
```

### Step 9. Group the entire dataframe by regiment and company

```python
grouped = regiment.groupby(['regiment','company'])
for (regiment,company),data in grouped:
  print(f"\nRegiment : {regiment}, Company : {company}")
  print(data)
```

### Step 10. What is the number of observations in each regiment and company

```python
observ_groub = regiment.groupby(['regiment', 'company'])
print(f"Number of observations in each regiment and company : {len(observ_groub)}")
```

### Step 11. Iterate over a group and print the name and the whole data from the regiment

```python
regiments_name = regiment.groupby('regiment')
for name,data in regiments_name:
  print(f"\nName : {name}")
  print(data)
```
