# US - Baby Names

### Introduction:

We are going to use a subset of [US Baby Names](https://www.kaggle.com/kaggle/us-baby-names) from Kaggle.  
In the file it will be names from 2004 until 2014


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv).

### Step 3. Assign it to a variable called baby_names.

```python
baby_names = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
baby_names
```

### Step 4. See the first 10 entries

```python
baby_names.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
baby_names.drop(['Unnamed: 0','Id'],axis=1,inplace=True)
baby_names
```

### Step 6. Is there more male or female names in the dataset?

```python
gender_count = baby_names['Gender'].value_counts()
pd.DataFrame(gender_count)
```

### Step 7. Group the dataset by name and assign to names

```python
names = baby_names.groupby('Name')
```

```python
for name,data in names:
  print(f"\nName {name}")
  print(data)
```

### Step 8. How many different names exist in the dataset?

```python
number_unique_names = baby_names['Name'].nunique()
print(f"Number different names exist in the dataset is : {number_unique_names}")
```

### Step 9. What is the name with most occurrences?

```python
most_occ_name = baby_names['Name'].value_counts().idxmax()
most_occ_name_number = baby_names['Name'].value_counts().max()
print(f"Name with most occurrences '{most_occ_name}' , with number '{most_occ_name_number}'")
```

### Step 10. How many different names have the least occurrences?

```python
least_occ_name = baby_names['Name'].value_counts().idxmin()
least_occ_name_number = baby_names['Name'].value_counts().min()
print(f"Name with least occurrences '{least_occ_name}' , with number '{least_occ_name_number}'")
```

### Step 11. What is the median name occurrence?

```python
median_name = baby_names['Name'].value_counts().median()
print(f"Median name occurrence : {median_name}")
```

### Step 12. What is the standard deviation of names?

```python
median_name = baby_names['Name'].value_counts().std()
print(f"Standard deviation of names : {median_name}")
```

### Step 13. Get a summary with the mean, min, max, std and quartiles.

```python
baby_names.describe()
```
