### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data)

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data",sep="\s+",parse_dates=[[0,1,2]])
data
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
date_range = pd.Series(pd.date_range(start='1/1/1961',end='12/31/1978',freq='D'))
data['Yr_Mo_Dy'] = date_range
data
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data.set_index('Yr_Mo_Dy',inplace=True)
data
```

### Step 6. Compute how many values are missing for each location over the entire record.  
#### They should be ignored in all calculations below.

```python
missing_number = data.isnull().sum()
print(f"Number values are missing : \n{missing_number}\n")
```

### Step 7. Compute how many non-missing values there are in total.

```python
non_missing = data.count().sum()-missing_number.sum()
print(f"Number of non_missing in total : {non_missing}")
```

### Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
#### A single number for the entire dataset.

```python
mean = data.mean()
print(f"Mean windspeeds of each location :\n{mean}")
```

```python
mean_windspeed_all_location = data.mean().mean()
print(f"Mean windspeeds over the location and all the times : {mean_windspeed_all_location}")
```

### Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days

#### A different set of numbers for each location.

```python
l = date_range.to_list()
loc_stats = pd.DataFrame()
for i in range(len(l)-1):
    stats = data.iloc[i,:].agg(['min', 'max', 'mean', 'std'])
    loc_stats = loc_stats.append(stats, ignore_index=True)

loc_stats
```

```python
loc_stats = pd.DataFrame({
    'min_windspeed': data.min(),
    'max_windspeed': data.max(),
    'mean_windspeed': data.mean(),
    'std_windspeed': data.std()
})
loc_stats
```

### Step 10. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

```python
data['Month_Year'] = data.index.to_period('M')
january_data = data[data.index.month == 1]
average_windspeed_january = january_data.groupby('Month_Year').mean()
average_windspeed_january[['RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']]
```
